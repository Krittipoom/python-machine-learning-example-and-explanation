# Principal Component Analysis (PCA)

## Import necessary libraries

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

## Load iris dataset

```python
data = load_iris()
```

```python
import pandas as pd
print(pd.DataFrame(data.data))
```

```python
print(pd.DataFrame(data.target))
```

## Create PCA object with 2 components

```python
pca = PCA(n_components=2)
```

## Transform data into principal components

```python
X_pca = pca.fit_transform(data.data)
```

## Plot transformed data

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c=data.target)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

This code loads the iris dataset, creates a PCA object with 2 components, transforms the data into its principal components, and plots the transformed data with colors corresponding to their original labels. You can run this code in a Jupyter Notebook or any other Python environment that has the necessary libraries installed.
