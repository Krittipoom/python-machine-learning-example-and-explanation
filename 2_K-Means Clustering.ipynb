# K-Means Clustering

## Import necessary libraries

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## Load iris dataset

```python
data = load_iris()
```

```python
import pandas as pd
print(pd.DataFrame(data.data))
```

```python
print(pd.DataFrame(data.target))
```

## Create k-means clustering object with 3 clusters

```python
kmeans = KMeans(n_clusters=3)
```

## Train k-means clustering on data

```python
kmeans.fit(data.data)
```

## Get cluster labels and cluster centers

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

## Plot data points with colors corresponding to their assigned clusters

```python
plt.scatter(data.data[:,0], data.data[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], marker='x', s=200, linewidths=3, color='r')
plt.show()
```

This code loads the iris dataset, creates a k-means clustering object with 3 clusters, trains the clustering on the data, and plots the data points with colors corresponding to their assigned clusters.
